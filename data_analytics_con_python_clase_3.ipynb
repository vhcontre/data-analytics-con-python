## üîç Scope (Alcance de las Variables)

El **scope** se refiere a la visibilidad y accesibilidad de las variables en diferentes partes de un programa.

### Tipos de Scope

| Tipo         | Descripci√≥n |
|--------------|-------------|
| **Local**    | Variables definidas dentro de una funci√≥n. Solo accesibles dentro de esa funci√≥n. |
| **Global**   | Variables definidas fuera de funciones. Accesibles desde cualquier parte del c√≥digo. |
| **No Local** | Variables en un √°mbito superior, no globales. Accesibles usando `nonlocal`. |

### üîÑ Resoluci√≥n del Scope

Python sigue el principio **LEGB**:

- **L**ocal  
- **E**nclosing (cerramiento)  
- **G**lobal  
- **B**uilt-in  

Este orden determina c√≥mo se busca una variable y en qu√© nivel se encuentra.

### üß™ Ejemplos

#### Scope Local

```python
def mi_funcion():
   x = 10
   print(x)

mi_funcion()  # Imprime 10
print(x)      # Error: x no est√° definido fuera de la funci√≥n
```

#### Scope Global

```python
y = 20

def otra_funcion():
   print(y)

otra_funcion()  # Imprime 20
```

#### Modificar variable global

```python
z = 30

def cambiar_global():
   global z
   z = 40

cambiar_global()
print(z)  # Imprime 40
```

#### Scope No Local

```python
def funcion_externa():
   a = 5
   def funcion_interna():
      nonlocal a
      a += 1
      print(a)

   funcion_interna()

funcion_externa()  # Imprime 6
```

---

## üìå Importancia del Scope

Comprender el scope ayuda a prevenir conflictos de nombres y errores de referencia.  
Usar correctamente `global` y `nonlocal` permite organizar el c√≥digo de forma clara y segura.

## Actividad 1: Tipo de Par√°metros

1. **Multiplicar:** Arranqu√© creando una funci√≥n muy simple que recibe dos n√∫meros y devuelve el producto. La prob√© con distintos pares de n√∫meros y funciona bien.
2. **Bienvenida:** Despu√©s hice una funci√≥n con un par√°metro por defecto. Si no pongo un mensaje, usa el que ya est√° definido; si quiero, puedo personalizarlo.
3. \**Promedio con *args:** Para el promedio us√© `*args`, que me deja pasar muchas notas sin tener que definir cu√°ntas. Prob√© con distintas cantidades y siempre calcula bien.
4. \*\*Concatena con **kwargs:** Por √∫ltimo, us√© `**kwargs`. Eso me permite armar informaci√≥n con pares clave-valor, como edad, ciudad, ocupaci√≥n. As√≠ puedo combinar datos de forma flexible.



```
# Tiene formato de c√≥digo
```




````python
# 1. Par√°metros Posicionales
def multiplicar(a, b):
    return a * b

# Aqu√≠ estoy creando una funci√≥n llamada "multiplicar" que recibe dos n√∫meros y devuelve su producto.
# La pruebo con algunos valores:
print("Ejemplo multiplicar:", multiplicar(3, 4))
print("Ejemplo multiplicar:", multiplicar(7, 2))


# 2. Par√°metros por Defecto
def bienvenida(nombre, mensaje="¬°Bienvenido!"):
    return f"{mensaje} {nombre}"

# En este caso defino un mensaje con valor por defecto.
# Si no escribo nada en el segundo argumento, Python usa "¬°Bienvenido!".
print(bienvenida("Alicia"))
print(bienvenida("V√≠ctor", "Qu√© alegr√≠a verte,"))


# 3. Par√°metros Indefinidos (*args)
def calcular_promedio(*notas):
    return sum(notas) / len(notas) if notas else 0

# Aqu√≠ prob√© con diferentes cantidades de notas, porque *args permite pasar muchos valores.
print("Promedio de 7, 8, 9:", calcular_promedio(7, 8, 9))
print("Promedio de 10, 5:", calcular_promedio(10, 5))


# 4. Par√°metros de Palabras Clave Indefinidas (**kwargs)
def concatena_info(nombre, **kwargs):
    info = f"Nombre: {nombre}"
    for clave, valor in kwargs.items():
        info += f", {clave}: {valor}"
    return info

# Ac√° uso **kwargs, que me deja pasar pares clave-valor como edad, ciudad, etc.
print(concatena_info("Alicia", edad=30, ciudad="C√≥rdoba", ocupaci√≥n="Estudiante"))
print(concatena_info("V√≠ctor", hobby="M√∫sica", lenguaje="Python"))
````
