## Actividad 2 Manipulaci√≥n de Datos

**Consigna:**

1. Obtener los nombres de las columnas.
2. Eliminar la columna `'deck'` (contiene nulos).
3. Reindexar el DataFrame.

```python
import pandas as pd
import seaborn as sns

# Cargar dataset Titanic usando Seaborn
df = sns.load_dataset('titanic')
print("Primeras filas del dataset cargado con Seaborn:")
display(df.head())

# Obtener nombres de las columnas
columnas = df.columns.tolist()
print("\nNombres de las columnas:")
print(columnas)

# Eliminar columna 'deck'
if 'deck' in df.columns:
    df = df.drop(columns=['deck'])
    print("\nColumna 'deck' eliminada correctamente.")
else:
    print("\nLa columna 'deck' no existe en el dataset.")

# Reindexar DataFrame
df = df.reset_index(drop=True)
print("\nDataFrame reindexado. Primeras filas del DataFrame modificado:")
display(df.head())

# Resumen final en tabla
resumen_final = pd.DataFrame({
    "N√∫mero de filas": [df.shape[0]],
    "N√∫mero de columnas": [df.shape[1]],
    "Columnas restantes": [df.columns.tolist()]
})
print("\nResumen final del DataFrame:")
display(resumen_final)
```
