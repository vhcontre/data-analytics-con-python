## Actividad 1: Agrupando Datos para Análisis de Ventas
Objetivo:
Aplicar la función groupby() de Pandas para segmentar los datos de ventas.

Calcular el total de ventas por producto.
Presentar los resultados de manera clara y concisa.
Datos: ventas.csv

```python
# -*- coding: utf-8 -*-
"""
Actividad 1: Agrupando Datos para Análisis de Ventas
Archivo: ventas.csv

Objetivo:
- Aplicar la función groupby() de Pandas para segmentar los datos de ventas.
- Calcular el total de ventas por producto.
- Presentar los resultados de manera clara y concisa.

Contexto:
Analizar un conjunto de datos sobre las ventas de productos en la última campaña.
Este informe será la base para decidir el enfoque de ventas en el próximo trimestre.
"""

# Importamos la librería necesaria
import pandas as pd

# 1️⃣ Cargar el dataset
df = pd.read_csv("ventas.csv")

# Revisamos los primeros registros para asegurarnos que se cargó correctamente
print("Primeros registros del dataset:")
print(df.head())

# 2️⃣ Agrupar las ventas por producto y sumar las ventas totales
# Usamos groupby() sobre la columna 'producto' y sumamos la columna 'ventas'
grupo_ventas = df.groupby('producto')['ventas'].sum().reset_index()

# Renombramos la columna de ventas para mayor claridad
grupo_ventas.rename(columns={'ventas': 'total_ventas'}, inplace=True)

# 3️⃣ Mostrar el DataFrame resultante con el total de ventas por producto
print("\nTotal de ventas por producto:")
print(grupo_ventas)

# 4️⃣ Extra: ordenar los resultados de mayor a menor para mejor visualización
grupo_ventas_ordenado = grupo_ventas.sort_values(by='total_ventas', ascending=False)
print("\nTotal de ventas por producto (ordenado de mayor a menor):")
print(grupo_ventas_ordenado)
```

## Actividad 2: Creación de Tablas Dinámicas para Resumir Datos
Objetivo:
Aplicar la función pivot_table() de Pandas para resumir información clave de los datos de ventas.
Crear un resumen que permita analizar las ventas mensuales por producto.
Interpretar los resultados y prepararlos para una presentación.

Contexto
Después de finalizar el análisis de ventas, ahora es el turno de Matías, tu mentor Data Analyst, de evaluar el rendimiento de las campañas de marketing. Matías te pide que utilices tablas dinámicas para resumir información sobre las ventas y el rendimiento de las campañas. Ejercicios Prácticos

- Cargar el conjunto de datos ventas.csv de la actividad anterior.
- Utilizar la función pivot_table() para crear una tabla dinámica que muestre las ventas mensuales por producto.
- Asegurarse de que la tabla muestre los productos en las filas y los meses en las columnas, con las ventas totales como valores dentro de la tabla.
-  Mostrar la tabla dinámica en pantalla.

```python
# -*- coding: utf-8 -*-
"""
Actividad 2: Creación de Tablas Dinámicas para Resumir Datos
Archivo: ventas.csv

Objetivo:
- Aplicar la función pivot_table() de Pandas para resumir información clave de los datos de ventas.
- Crear un resumen que permita analizar las ventas mensuales por producto.
- Interpretar los resultados y prepararlos para una presentación.

Contexto:
Evaluar el rendimiento de las campañas de marketing usando tablas dinámicas.
Esta información ayudará a entender qué estrategias funcionan mejor y dónde se pueden hacer mejoras.
"""

# Importamos la librería necesaria
import pandas as pd

# 1️⃣ Cargar el dataset de la actividad anterior
df = pd.read_csv("ventas.csv")

# Revisamos los primeros registros para asegurarnos que se cargó correctamente
print("Primeros registros del dataset:")
print(df.head())

# 2️⃣ Crear la tabla dinámica con pivot_table()
# Filas: 'producto'
# Columnas: 'mes'
# Valores: 'ventas', agregando con suma
tabla_dinamica = pd.pivot_table(
    df,
    index='producto',
    columns='mes',
    values='ventas',
    aggfunc='sum',
    fill_value=0  # Para reemplazar NaN con 0
)

# 3️⃣ Mostrar la tabla dinámica en pantalla
print("\nTabla dinámica: ventas mensuales por producto")
print(tabla_dinamica)

# 4️⃣ Extra: ordenar las filas alfabéticamente y las columnas por meses (opcional)
meses_orden = ['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
tabla_dinamica = tabla_dinamica.reindex(columns=meses_orden, fill_value=0)
tabla_dinamica = tabla_dinamica.sort_index()

print("\nTabla dinámica ordenada:")
print(tabla_dinamica)
```

## Salida Visual

```python
# -*- coding: utf-8 -*-
"""
Actividad 2 (Visualización): Gráfico de Ventas Mensuales por Producto
Archivo: ventas.csv

Objetivo:
- Visualizar las ventas mensuales por producto usando un gráfico de barras agrupadas.
- Facilitar la interpretación de los resultados y preparar la presentación.
"""

# Importamos librerías necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1️⃣ Cargar el dataset
df = pd.read_csv("ventas.csv")

# 2️⃣ Crear la tabla dinámica como en la actividad anterior
tabla_dinamica = pd.pivot_table(
    df,
    index='producto',
    columns='mes',
    values='ventas',
    aggfunc='sum',
    fill_value=0
)

# Ordenar columnas por meses
meses_orden = ['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
tabla_dinamica = tabla_dinamica.reindex(columns=meses_orden)

# 3️⃣ Preparar el DataFrame para Seaborn (long format)
tabla_long = tabla_dinamica.reset_index().melt(id_vars='producto', var_name='Mes', value_name='Ventas')

# 4️⃣ Crear gráfico de barras agrupadas
plt.figure(figsize=(14,7))
sns.barplot(data=tabla_long, x='Mes', y='Ventas', hue='producto')
plt.title('Ventas Mensuales por Producto', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ventas', fontsize=12)
plt.legend(title='Producto')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
